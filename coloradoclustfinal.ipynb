```python
data = pd.read_csv("/kaggle/input/wine-dataset-for-clustering/wine-clustering.csv")

print(data)

print(data.columns)
```

```python
data =data.dropna(axis=0)
```

Some preprocessing

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
datapr = StandardScaler().fit_transform(data)
pca = PCA().fit(datapr)
datapr = pca.transform(datapr)
```

EDA:finding most important features

```python
import numpy as np

# Get loadings: shape (n_components, n_features)
loadings = pca.components_

# Sum absolute weights across top k components (e.g. first 3)
k = 3
feature_scores = np.sum(np.abs(loadings[:k, :]), axis=0)

# Get feature names and sort
important_features = np.array(data.columns)[np.argsort(feature_scores)[::-1][:3]]
print("Top 3 important features:", important_features)
```

Main algorithm 

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = model.fit_predict(data)

print(labels)
```
